```python
# 创建文件
def mkdir(path):
    # 输入文件路径
    folder = os.path.exists(path)

    if not folder:  # 判断是否存在文件夹如果不存在则创建为文件夹
        os.makedirs(path)  # makedirs 创建文件时如果路径不存在会创建这个路径
        print("---  new folder...  ---")
        print("---  OK  ---")

    else:
        print("---  There is this folder!  ---")
# 根据时间对数据进行切片

def cutDataByTime(data, StartTime, EndTime, tf, tt):
    '''data:传入的结果数据,StartTime:起始时间,EndTime:结束时间,tt:交易期长度'''
    dataCopy = data.copy()
    dataCopy['name'] = pd.to_datetime(dataCopy['name'])
    StartTime = datetime.datetime.strptime(StartTime, "%Y-%m-%d")
    EndTime = datetime.datetime.strptime(EndTime, "%Y-%m-%d")
    EndTime = EndTime - relativedelta(months=tt) - relativedelta(months=tf)
    dataCopy = dataCopy[(dataCopy['name'] >= StartTime) & (dataCopy['name'] <= EndTime)]
    return dataCopy


# 根据时间切片计算汇总结果
def outputBestTTAndTF(kli):
    '''IDSet:id集合,IsSH:是否沪股，startTime:起始时间,endTime:结束时间'''
    All = 1000000
    CD = 0.05
    mkdir(pathlib.Path('..', 'BestTTAndTF'))
    path1 = None
    path2 = None
    path3 = None
    earlyTime = '2002-04-02'
    endTime = '2023-08-25'
    endTime = datetime.datetime.strptime(endTime, "%Y-%m-%d")
    earlyTime = datetime.datetime.strptime(earlyTime, "%Y-%m-%d")
    OutputStartTime = ''
    OutputEndTime = ''
    ouputdf = pd.DataFrame(np.zeros((24, 12)))
    countdf = pd.DataFrame(np.zeros((24, 12)))
    SHdatapath1 = pathlib.Path('..', 'SummaryByTimeFixed1', 'SH')
    # 设置沪港通前后的时间范围
    SHOutputStartTime = '2015-05-01'
    SHOutputEndTime = '2023-08-01'
    SZdatapath1 = pathlib.Path('..', 'SummaryByTimeFixed1', 'SZ')
    # 设置深港通前后的时间范围
    SZOutputStartTime = '2017-06-01'
    SZOutputEndTime = '2023-08-01'
    # kli = np.array([0, 0.25, 0.5, 0.75, 1])
    # [0,0.25,0.5,0.75,1]
    # [0.05,0.10,0.15,0.2]
    # [0.01,0.02,0.03,0.04]
    # [0.002,0.004,0.006,0.008]
    # [0.0004,0.0008,0.0012,0.0016]
    # [0.0001,0.0002,0.003,0.004]
    # [0.000025,0.00005,0.000075]
    # [0.000005,0.00001,0.000015,0.00002]
    # [0.000001,0.000002,0.000003,0.000004]
    # 0.0001,0.005,0.015,0.02
    for k in kli:
        print('k=', k)
        SHdatapath2 = SHdatapath1.joinpath('k' + str(k))
        SZdatapath2 = SZdatapath1.joinpath('k' + str(k))
        for tf in range(1, 25):
            for tt in range(1, 13):
        # for tf in range(1, 3):
        #     for tt in range(1, 3):
                # for tf in range(10, 11):
                #     for tt in range(12, 13):
                print('tf=', tf, 'tt=', tt)
                SHdatapath3 = SHdatapath2.joinpath('tf' + str(tf) + 'tt' + str(tt),'SH'+'k'+str(k)+'tf'+str(tf)+'tt'+str(tt)+'PostSummary.csv')
                SZdatapath3 = SZdatapath2.joinpath('tf' + str(tf) + 'tt' + str(tt),'SZ'+'k'+str(k)+'tf'+str(tf)+'tt'+str(tt)+'PostSummary.csv')
                SHsummary=pd.read_csv(SHdatapath3)
                SZsummary=pd.read_csv(SZdatapath3)
                ouputdf.iat[tf-1, tt-1] += SHsummary["annualYieldRate(non-compounding)"].mean()
                countdf.iat[tf-1,tt-1]+=1
                ouputdf.iat[tf-1, tt-1] += SZsummary["annualYieldRate(non-compounding)"].mean()
                countdf.iat[tf-1,tt-1]+=1
                print('tf' + str(tf) + 'tt' + str(tt) + ' end')
        print('k=', k, ' end')
    outputdf=ouputdf/countdf
    # if IsSH:
    #     DataOutPath =pathlib.Path('..', 'BestTTAndTF','SHSummary.csv')
    # else:
    #     DataOutPath =pathlib.Path('..', 'BestTTAndTF','SZSummary.csv')
    DataOutPath =pathlib.Path('..', 'BestTTAndTF','Summary.csv')
    outputdf.to_csv(DataOutPath)
    print("*" * 100)
    print("程序结束")


kliTrade = np.array([0,0.25,0.5,0.75,1,1.25,1.5,1.75,2])
# kliTrade = np.array([0,0.25,0.5,0.75,1])
outputBestTTAndTF(kliTrade)
```
